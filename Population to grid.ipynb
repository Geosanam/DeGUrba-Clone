```python
joined_grid_muni.set_geometry('geometry',inplace=True)
```

```python
joined_grid_muni.rename(columns={'index_right':'muni_index'},inplace=True)
```

```python
joined_grid_muni
```

```python
joined_grid_buildings = joined_grid_muni.sjoin(gdf_buildings,how='left',predicate='intersects')
```

```python
grid_builds = joined_grid_buildings.groupby(['id']).agg({
    'type':'count'
}).reset_index()
```

```python
grid_builds.rename(columns={'type':'builds_grid'},inplace=True)
```

```python
grid_builds_count = joined_grid_muni.merge(grid_builds,on='id',how='left')
```

```python
grid_builds_count
```

```python
grid_builds_count['grid_popn'] = (grid_builds_count['builds_grid']/grid_builds_count['builds_cou'])*grid_builds_count['Pop_int']
```

```python
grid_builds_count[grid_builds_count['name']=='Kathmandu Metropolitian City']['grid_popn'].sum()
```

```python
grid_builds_count[['id','muni_index','name','Pop_int','builds_cou','builds_grid','grid_popn','geometry']].to_file('../Data/Degurba data/PopGridNepal_final.shp')
```
