```python
# This returns the overlapped area

def overlapped_area(row):
    loc_column = row['PrimaryKey']
    
    # Get the geometry of the row
    row_geometry = row['geometry']
    
    # Find the urban geometry for the given loc_column
    urban_geometry = gdf_urban[gdf_urban['PrimaryKey'] == loc_column].geometry.iloc[0]
    
    # Check if the urban_geometry is empty (no match found for 'PrimaryKey')
    # if urban_geometry.is_empty:
    #     return float('nan')
    
    # Calculate the area of the intersection
    overlapped_area = row_geometry.intersection(urban_geometry).area
    
    return overlapped_area
```

```python
joined['overlapped_area'] = joined.apply(overlapped_area,axis=1)
```

```python
joined[['name','Pop_int','Class','Population_right','overlapped_area']].sort_values(by='name').head(20)
```

```python
def overlapped_geom(row):
    loc_column = row['PrimaryKey']
    
    # Get the geometry of the row
    row_geometry = row['geometry']
    
    # Find the urban geometry for the given loc_column
    urban_geometry = gdf_urban[gdf_urban['PrimaryKey'] == loc_column].geometry.iloc[0]
    
    # Check if the urban_geometry is empty (no match found for 'PrimaryKey')
    # if urban_geometry.is_empty:
    #     return float('nan')
    
    # Calculate the area of the intersection
    overlapped = row_geometry.intersection(urban_geometry)   #geometry
    
    return overlapped
```

```python
joined['overlapped_geom'] = joined.apply(overlapped_geom,axis=1)
```

```python
joined_sorted = joined[['name','Pop_int','Class','label','Population_right','overlapped_area','Area']].sort_values(by='name')
```

```python
joined_sorted.head()
```

```python
joined_sorted['overlapped_area_ratio'] = joined_sorted['overlapped_area']/joined_sorted['Area']
```

```python
joined_sorted.head()
```

```python
joined_sorted['overlapped_pop'] = joined_sorted['overlapped_area_ratio'] * joined_sorted['Population_right']
```

```python
# Define cities
cities = joined_sorted_overlapped[(joined_sorted_overlapped['Class']=='Centre') & (joined_sorted_overlapped['overlapped_pop']>=joined_sorted_overlapped['Pop_int']/2)]
```

```python
cities
```

```python
# Define towns

initial_criteria = (
    (joined_sorted_overlapped['Class'] == 'Centre')  # Urban centers
    & (joined_sorted_overlapped['overlapped_pop'] < joined_sorted_overlapped['Pop_int']/2)  # Less than 50% in urban centers
)

selected_names = joined_sorted_overlapped.loc[initial_criteria, 'name'].unique() # Find out those units which fulfill the first criteria to be used in second criteria

# final_criteria = (
#     (joined_sorted_overlapped['name'].isin(selected_names))  # Names from the initial criteria
#     & (joined_sorted_overlapped['Class'] == 'Rural')  # classified as 'Centre'
#     & (joined_sorted_overlapped['overlapped_pop'] <= joined_sorted_overlapped['Pop_int']/2)  # No more than 50% in rural
# )

towns = joined_sorted_overlapped[initial_criteria]
```

```python
towns
```

```python
# Create a column to mark down the spatial units in Degurba class

cities['Degurba_class'] = 1
towns['Degurba_class'] = 2
```

```python
import pandas as pd
```

```python
# The rural are the area other than towns and cities. Therefore, the towns and cities are initially joined which will be filtered out from the overall data

cities_towns =pd.concat([cities, towns], ignore_index=True)
```

```python
# This filters out the towns and cities from joined_sorted_overlapped and result only the remaining

rural = joined_sorted_overlapped[~joined_sorted_overlapped['name'].isin(cities_towns['name'].unique())]
```

```python
rural
```

```python
rural['Degurba_class'] = 3
```

```python
# Finally, all are joined
# The resultant df(not a gdf) can be merged(joined) with the gdf of spatial units to classify them as degurba classes
 
cities_towns_rural = pd.concat([cities_towns, rural], ignore_index=True)
```

```python
# The columns to join while applying the join can be specified as follows

final_classification = gdf_pop.merge(cities_towns_rural[['name','Degurba_class']],on='name',how='inner')
```
